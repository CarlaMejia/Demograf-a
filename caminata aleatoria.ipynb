# Caminata Simulación

Para realizar este análisis realizaremos una simulación de una camioneta en python, teniendo solo dos alternativas

1. caminar un paso hacia adelante
2. caminar un paso hacia atrás

A continuación se presenta la simulación

```python
#Importamos librerias de numpy y matplotlib
import numpy as np
import matplotlib.pyplot as plt
#Inicializamos los valores
posicion=0 #ponemos el cero sin lista para que podamos sumarle un número, en lista no es posible
caminata=[posicion]#agregamos el valor anterior a la lista
n_pasos=100
#Inicio del ciclo
for i in range(n_pasos): #in range(100)
    numero_aleatorio=np.random.randint(0,2)
    if numero_aleatorio==1:
        paso=-1
    else: 
        paso=1
    posicion= posicion + paso
    caminata.append(posicion)
print(caminata[:20])
##graficar
plt.plot(caminata)
    
```
